```python
import numpy as np
import matplotlib.pyplot as plt
```

# Fun√ßoes

```python
def create_grid(N):
    grid = np.zeros((N, N))
    first_infected = np.random.randint(0, N), np.random.randint(0, N)
    grid[first_infected] = 1
    return grid
```

```python
def iteration(grid, recuperacao_prob, susceptibilidade_prob):
    N = grid.shape[0]
    new_grid = np.zeros((N, N))
    for i in range(N):
        for j in range(N):
            if grid[i, j] == 2:
                new_grid[i, j] = 2
            elif grid[i, j] == 1:
                if np.random.rand() < recuperacao_prob:
                    new_grid[i, j] = 2
                else:
                    new_grid[i, j] = 1
                if grid[(i + 1) % N, j] == 0 and np.random.rand() < susceptibilidade_prob:
                    new_grid[(i + 1) % N, j] = 1
                if grid[(i - 1), j] == 0 and np.random.rand() < susceptibilidade_prob:
                    new_grid[(i - 1), j] = 1
                if grid[i, (j + 1) % N] == 0 and np.random.rand() < susceptibilidade_prob:
                    new_grid[i, (j + 1) % N] = 1
                if grid[i, (j - 1)] == 0 and np.random.rand() < susceptibilidade_prob:
                    new_grid[i, (j - 1)] = 1
    return new_grid
```

```python
def count_values(grid):
    return [np.sum(grid == 0), np.sum(grid == 1), np.sum(grid == 2)]
```

```python
import matplotlib.colors as mcolors

def plot_grid(grid, counts, ax):
    colors = ['black', 'red', 'yellow']
    cmap = mcolors.ListedColormap(colors)
    bounds = [-0.5, 0.5, 1.5, 2.5]
    norm = mcolors.BoundaryNorm(bounds, cmap.N)

    ax.imshow(grid, cmap=cmap, norm=norm)
    ax.set_title('Infected: {}, Recovered: {}, Susceptible: {}'.format(counts[1], counts[2], counts[0]))
    ax.axis('off')
```

```python
def simulate_grid_evolution(create_grid, iteration, count_values, plot_grid, 
                            grid_size=40, n_steps=500, recuperacao_prob=0.1, 
                            susceptibilidade_prob=0.1, num_subplots=10):
    fig, ax = plt.subplots(2, 5, figsize=(30, 15))
    ax = ax.flatten()

    grid = create_grid(grid_size)
    contador = []

    for i in range(n_steps):
        grid = iteration(grid, recuperacao_prob, susceptibilidade_prob)
        counts = count_values(grid)
        contador.append(counts)
        
        if (i % (n_steps // num_subplots) == 0) and (i // (n_steps // num_subplots) < num_subplots):
            plot_grid(grid, counts, ax[i // (n_steps // num_subplots)])
        
        # if np.all(grid == 2):
        #     print("Simulation ended early: all grid elements are in state 2.")
        #     break

    plt.tight_layout()
    plt.show()

    return contador
```
