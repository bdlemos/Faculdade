```python
#Faça um programa em python que leia numeros binarios de um arquivo e plote eles em um grafico

import matplotlib.pyplot as plt

# Defina o nome do arquivo que contém os números binários
nome_arquivo = "n3.txt"

# Leia os números binários do arquivo
with open(nome_arquivo, "r") as arquivo:
    numeros_binarios = arquivo.readlines()

# Limpe os caracteres de quebra de linha (\n) e converta os números binários em inteiros
numeros_decimais = [int(binario.strip(), 2) for binario in numeros_binarios]

# Crie uma lista de índices para o eixo x
indices = list(range(len(numeros_decimais)))

# Plote os números no gráfico
plt.plot(indices, numeros_decimais, marker='o')
plt.xlabel('Ciclo de clock')
plt.ylabel('Número Decimal')
plt.title('x³ + x² + 1')
plt.grid(True)
plt.show()
```

```python
# import LFSR
#CRIA O POLINOMIO DE CODIFICAÇÃO E ARMAZENA OS BITS DE SAIDA EM UM ARQUIVO
import numpy as np
from pylfsr import LFSR

L = LFSR(fpoly=[16,14,13,11])

for _ in range(100000):
    #print(L.count,L.state,'',L.outbit,L.seq,sep='\t')
    L.next()
print('-'*50)
#escrita no arquivo a sequencia de saida
with open('pol16.txt', 'w') as f:
    for item in L.seq:
       f.write("%s\n" % item)    
print('Output: ',L.seq)
```

```python
from PIL import Image
#LÊ O ARQUIVO DE SAIDA E PLOTA EM UMA IMAGEM
with open("pol16.txt") as f:
    lines = f.read().splitlines ()
    img = Image.open("grafico2d.jpeg")
    print(img.size)
    #img.show()
    width = img.size [0]
    height = img.size [1]
    um = 0
    zero = 0
    for i in range(0, width):
        for j in range(0, height):
            data = img.getpixel ((i,j))
            if lines[i*width+j] == "0":
                img.putpixel ((i,j) ,(0,0,0))
                zero += 1
            else:
                img.putpixel ((i,j) ,(255,255,255))
                um += 1
    img.show()
    img.save("pol16.png")
    print("Zero/Total: ",zero/(zero+um))
    print("Um/Total: ",um/(zero+um))
    
```

```python
from PIL import Image
#LE UMA IMAGEM E CODIFICA COM BASE NA IMAGEM GERADA NO MODULO ANTERIOR
with open("pol16.txt") as f:
    lines = f.read().splitlines ()
    img = Image.open("grafico2d.jpeg")
    bitmap = Image.open("pol16.png")
    print(img.size)
    img.show()
    #bitmap.show()
    width = img.size [0]
    height = img.size [1]
    for i in range(0, width):
        for j in range(0, height):
            data1 = img.getpixel ((i,j))
            data2 = bitmap.getpixel ((i,j))
            #arredondamento para 0 ou 1
            if data1 != (255,255,255) and data1 != (0,0,0):
                media = (data1[0]+data1[1]+data1[2])/3
                if media > 127:
                    data1 = (255,255,255)
                else:
                    data1 = (0,0,0)
            if data1 != data2:
                img.putpixel ((i,j) ,(255,255,255))
            else:
                img.putpixel ((i,j) ,(0,0,0))
    img.show()
    img.save("codificadaPol16.png")
```

```python
from PIL import Image
#DECODIFICA A IMAGEM CODIFICADA
img = Image.open("pol16.png")
bitmap = Image.open("codificadaPol16.png")
print(img.size)
#img.show()
bitmap.show()
width = img.size [0]
height = img.size [1]
for i in range(0, width):
    for j in range(0, height):
        data1 = img.getpixel ((i,j))
        data2 = bitmap.getpixel ((i,j))
        if data1 != data2:
            img.putpixel ((i,j) ,(255,255,255))
        else:
            img.putpixel ((i,j) ,(0,0,0))
img.show()
```
