```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_csv_files(csv_files):
    for csv_file in csv_files:
        df = pd.read_csv(csv_file)  # Lê o arquivo CSV e cria um DataFrame
        csv_file = csv_file[:-4]
        df.sort_values(by=['N'], inplace=True)  # Ordena o DataFrame pelo número de pontos (N
        plt.plot(df['N'], df['Tempo'], label=csv_file)  # Plota os dados do arquivo CSV

    plt.ylabel('Tempo(ms)')  # Define o rótulo do eixo x
    plt.xlabel('Tamanho da entrada em caracteres')  # Define o rótulo do eixo y
    plt.title('Gráfico de Tempos de Compressão e Descompressão')  # Define o título do gráfico
    plt.legend()  # Adiciona a legenda

    plt.show()  # Exibe o gráfico


# Lista com os nomes dos arquivos CSV
csv_files = ['Compressão.csv', 'Descompressão.csv']

plot_csv_files(csv_files)
```
