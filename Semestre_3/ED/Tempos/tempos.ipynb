```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_csv_files(csv_files):
    for csv_file in csv_files:
        df = pd.read_csv(csv_file)  # L o arquivo CSV e cria um DataFrame
        csv_file = csv_file[:-4]
        df.sort_values(by=['N'], inplace=True)  # Ordena o DataFrame pelo nmero de pontos (N
        plt.plot(df['N'], df['Tempo'], label=csv_file)  # Plota os dados do arquivo CSV

    plt.ylabel('Tempo(ms)')  # Define o rtulo do eixo x
    plt.xlabel('Tamanho da entrada em caracteres')  # Define o rtulo do eixo y
    plt.title('Grfico de Tempos de Compresso e Descompresso')  # Define o ttulo do grfico
    plt.legend()  # Adiciona a legenda

    plt.show()  # Exibe o grfico


# Lista com os nomes dos arquivos CSV
csv_files = ['Compresso.csv', 'Descompresso.csv']

plot_csv_files(csv_files)
```
